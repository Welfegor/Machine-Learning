```python
# Импорт библиотек и загрузка данных
from sklearn.datasets import load_wine
import pandas as pd
import statsmodels.formula.api as sm
from sklearn.metrics import mean_squared_error, median_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wine_data = load_wine()
df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
df['target'] = wine_data.target
display(df)
```

```python
# Исключение регионов с индексом 2
df = df[df.target != 2]
display(df)
```

```python
# Выявление самых коррелирующих с 'target' параметров
plt.figure(figsize=(15, 13), dpi=80)
sns.heatmap(df.loc[:, 'alcohol':'target'].corr(), xticklabels=df.loc[:, 'alcohol':'target'].corr().columns, center=0, annot=True)
plt.show()
```

```python
# Построение логистической регрессии по отобранным параметрам
selected_features = ['alcohol', 'color_intensity', 'proline']
logistic_model = sm.logit('target ~ alcohol + color_intensity + proline', df)
results_logistic = logistic_model.fit()
print(results_logistic.summary())
```

```python
# Вычисление метрик качества классификации
y_true = df.loc[:, 'target']
y_pred = results_logistic.predict(df.loc[:, selected_features])
display(y_pred)
```

```python
# Используем метрики MSE и MAE
errors = {'MSE': [mean_squared_error(y_true, y_pred)],
          'MAE': [median_absolute_error(y_true, y_pred)]}

display(pd.DataFrame(data=errors))
```
