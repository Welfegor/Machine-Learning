```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
import numpy as np
```

```python
# Загрузка и предобработка данных
file_path = 'Boston.csv'
dataset = pd.read_csv(file_path, sep='[,]', engine='python')
df = dataset.copy()
display(df)
```

```python
# Подготовка датасета к работе
df.columns = df.columns.str.replace(r'\W', '', regex=True)
df = df.drop(df.columns[[0]], axis=1)
display(df)
```

```python
# Построение таблицы корреляции
plt.figure(figsize=(15, 13), dpi=80)
sns.heatmap(df.loc[:, 'crim':'medv'].corr(), xticklabels=df.loc[:, 'crim':'medv'].corr().columns, center=0, annot=True)
plt.show()
```

```python
# 'lstat' - самый коррелирующий признак, построим график зависимости
plt.plot(df['lstat'], df['medv'], 'ro', markersize=2)
plt.xlabel('lstat')
plt.ylabel('medv')
plt.title('Зависимость стоимости жилья от количества населения')
plt.show()
```

```python
# Построение модели линейной регрессии с параметром 'lstat'
model_lstat = sm.ols('medv ~ lstat', df)
results_lstat = model_lstat.fit()
print(results_lstat.summary())
```

```python
# Построение графика регрессии
b, a = np.polyfit(df['lstat'], df['medv'], deg=1)
xseq = np.linspace(0, 38)

plt.plot(df['lstat'], df['medv'], 'ro', markersize=2)
plt.plot(xseq, a + b * xseq, color='k', lw=2.5)
plt.xlabel('lstat')
plt.ylabel('medv')
plt.title('Зависимость стоимости жилья от количества населения')
plt.show()
```

```python
# Построение модели линейной регрессии со всеми параметрами
model_all_params = sm.ols('medv ~ crim + zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + black + lstat', df)
results_all_params = model_all_params.fit()
print(results_all_params.summary())
```
